# Creating Data for Julia DS Practice

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# generate data that's more or less y ~ 2*x1 + 5*x2 + 4
x1 = np.random.uniform(-4., 5., 100)
x2 = np.random.uniform(-5., 3., 100)

sig = np.random.normal(100)

b0 = 4.; b1 = 2.; b2 = 5.
y = b0 + b1*x1 + b2*x2 + sig
```

```python
# create dataframe
df = pd.DataFrame({'x1': x1, 'x2': x2, 'y': y})
df.to_csv('../data/sample_data.csv')
```

```python
# plot
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(df['x1'], df['x2'], df['y'])

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
plt.tight_layout()
plt.show()
```
